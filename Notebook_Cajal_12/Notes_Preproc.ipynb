```python
%matplotlib inline
import nilearn
```

```python
from nilearn import datasets
```

```python
# Note that the variable MNI152_FILE_PATH is just a path to a Nifti file
print('Path to MNI152 template: {}'.format(datasets.MNI152_FILE_PATH))
```

```python
from nilearn import plotting
plotting.plot_img(datasets.MNI152_FILE_PATH)
```

```python
from nilearn import image
smooth_anat_img = image.smooth_img(datasets.MNI152_FILE_PATH, fwhm=20)
```

```python
print(smooth_anat_img)
```

```python
plotting.plot_img(smooth_anat_img)
```

```python
smooth_anat_img2 = image.smooth_img(datasets.MNI152_FILE_PATH, fwhm=2)
plotting.plot_img(smooth_anat_img2)
```

```python
def smooth_plot(kernelSize, vis=True):
    smooth_anat_img = image.smooth_img(datasets.MNI152_FILE_PATH, fwhm=kernelSize)
    plotting.plot_img(smooth_anat_img,cut_coords=(-20,4,31))
    if vis:
        plotting.plot_img(smooth_anat_img,cut_coords=(-20,4,31),output_file='testfile_%i.png' % kernelSize)
    return smooth_anat_img
```

```python
smooth_plot(5)
```
